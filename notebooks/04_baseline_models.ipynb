```python
#Locate project root and load saved artifacts

from pathlib import Path
import json
import pandas as pd
import sklearn

# 1. Locate the project root automatically

CURRENT_DIRECTORY = Path.cwd().resolve()

# Search the current directory and its parent directories
# for the main project structure.
candidate_directories = [
    CURRENT_DIRECTORY,
    *CURRENT_DIRECTORY.parents
]

PROJECT_ROOT = None

for candidate in candidate_directories:
    dataset_candidate = (
        candidate
        / "data"
        / "processed"
        / "diabetic_modeling_data_final.csv"
    )

    if dataset_candidate.exists():
        PROJECT_ROOT = candidate
        break

if PROJECT_ROOT is None:
    raise FileNotFoundError(
        "Could not locate the hospital readmission project root.\n"
        f"Current working directory: {CURRENT_DIRECTORY}"
    )

print(f"Current working directory: {CURRENT_DIRECTORY}")
print(f"Project root detected: {PROJECT_ROOT}")

# 2. Define absolute project file paths

DATA_PATH = (
    PROJECT_ROOT
    / "data"
    / "processed"
    / "diabetic_modeling_data_final.csv"
)

SCHEMA_PATH = (
    PROJECT_ROOT
    / "data"
    / "processed"
    / "model_feature_schema.json"
)

SPLIT_PATH = (
    PROJECT_ROOT
    / "data"
    / "processed"
    / "patient_split_assignments.csv"
)

PREPROCESSING_METADATA_PATH = (
    PROJECT_ROOT
    / "artifacts"
    / "preprocessing_metadata.json"
)

required_files = {
    "Final modeling dataset": DATA_PATH,
    "Feature schema": SCHEMA_PATH,
    "Patient split assignments": SPLIT_PATH,
    "Preprocessing metadata": PREPROCESSING_METADATA_PATH,
}

# 3. Confirm that all required files exist

missing_files = [
    f"{name}: {path}"
    for name, path in required_files.items()
    if not path.exists()
]

if missing_files:
    raise FileNotFoundError(
        "The following required files were not found:\n"
        + "\n".join(missing_files)
    )

print("\nAll required Notebook 4 input files were found.")

# 4. Load saved project artifacts

df = pd.read_csv(DATA_PATH)

split_assignments = pd.read_csv(SPLIT_PATH)

with open(SCHEMA_PATH, "r", encoding="utf-8") as file:
    feature_schema = json.load(file)

with open(
    PREPROCESSING_METADATA_PATH,
    "r",
    encoding="utf-8"
) as file:
    preprocessing_metadata = json.load(file)

# 5. Display loading information


print(f"Pandas version: {pd.__version__}")
print(f"Scikit-learn version: {sklearn.__version__}")

print("\nFinal modeling dataset:")
print(f"Shape: {df.shape}")
print(f"Missing values: {df.isna().sum().sum():,}")

print("\nPatient split assignments:")
print(f"Shape: {split_assignments.shape}")
print(f"Columns: {split_assignments.columns.tolist()}")

print("\nFeature schema:")
print(f"Top-level keys: {list(feature_schema.keys())}")

print("\nPreprocessing metadata:")
print(f"Top-level keys: {list(preprocessing_metadata.keys())}")


# Step 2: Validate saved splits and create fixed partitions

# 1. Read modeling definitions from the saved feature schema

TARGET_COLUMN = feature_schema["target_column"]
GROUP_COLUMN = feature_schema["group_split_column"]
IDENTIFIER_COLUMNS = feature_schema["identifier_columns"]

NUMERIC_FEATURES = feature_schema["numeric_model_features"]
CATEGORICAL_FEATURES = feature_schema["categorical_model_features"]
MODEL_FEATURES = feature_schema["all_model_features"]

# 2. Validate the feature definitions

if len(NUMERIC_FEATURES) != 8:
    raise ValueError(
        f"Expected 8 numeric predictors, found {len(NUMERIC_FEATURES)}."
    )

if len(CATEGORICAL_FEATURES) != 35:
    raise ValueError(
        f"Expected 35 categorical predictors, "
        f"found {len(CATEGORICAL_FEATURES)}."
    )

if len(MODEL_FEATURES) != 43:
    raise ValueError(
        f"Expected 43 total predictors, found {len(MODEL_FEATURES)}."
    )

if len(MODEL_FEATURES) != len(set(MODEL_FEATURES)):
    raise ValueError("Duplicate feature names were found in MODEL_FEATURES.")

feature_overlap = set(NUMERIC_FEATURES).intersection(
    CATEGORICAL_FEATURES
)

if feature_overlap:
    raise ValueError(
        "Some features appear in both numeric and categorical lists: "
        f"{sorted(feature_overlap)}"
    )

required_dataset_columns = set(
    MODEL_FEATURES
    + IDENTIFIER_COLUMNS
    + [TARGET_COLUMN, GROUP_COLUMN]
)

missing_dataset_columns = (
    required_dataset_columns - set(df.columns)
)

if missing_dataset_columns:
    raise ValueError(
        "Required columns are missing from the modeling dataset: "
        f"{sorted(missing_dataset_columns)}"
    )

# Prevent identifiers and target from entering the predictors
for excluded_column in [
    TARGET_COLUMN,
    GROUP_COLUMN,
    "encounter_id",
]:
    if excluded_column in MODEL_FEATURES:
        raise ValueError(
            f"{excluded_column} must not be included as a predictor."
        )

# 3. Validate the saved split-assignment file

required_assignment_columns = {
    "encounter_id",
    GROUP_COLUMN,
    TARGET_COLUMN,
    "split",
}

missing_assignment_columns = (
    required_assignment_columns - set(split_assignments.columns)
)

if missing_assignment_columns:
    raise ValueError(
        "Required columns are missing from split assignments: "
        f"{sorted(missing_assignment_columns)}"
    )

if df["encounter_id"].duplicated().any():
    raise ValueError(
        "Duplicate encounter_id values were found in the dataset."
    )

if split_assignments["encounter_id"].duplicated().any():
    raise ValueError(
        "Duplicate encounter_id values were found in split assignments."
    )

expected_split_labels = {"train", "validation", "test"}
actual_split_labels = set(
    split_assignments["split"].dropna().unique()
)

if actual_split_labels != expected_split_labels:
    raise ValueError(
        "Unexpected split labels found. "
        f"Expected {expected_split_labels}, "
        f"but found {actual_split_labels}."
    )

# 4. Confirm that dataset rows match split-assignment rows

integrity_check = df[
    ["encounter_id", GROUP_COLUMN, TARGET_COLUMN]
].merge(
    split_assignments[
        ["encounter_id", GROUP_COLUMN, TARGET_COLUMN, "split"]
    ],
    on="encounter_id",
    how="outer",
    suffixes=("_dataset", "_assignment"),
    indicator=True,
    validate="one_to_one",
)

if not integrity_check["_merge"].eq("both").all():
    unmatched_counts = integrity_check["_merge"].value_counts()
    raise ValueError(
        "The dataset and split assignments do not contain "
        "the same encounters:\n"
        f"{unmatched_counts}"
    )

group_match = (
    integrity_check[f"{GROUP_COLUMN}_dataset"]
    == integrity_check[f"{GROUP_COLUMN}_assignment"]
)

if not group_match.all():
    raise ValueError(
        "patient_nbr values do not match between the dataset "
        "and split assignments."
    )

target_match = (
    integrity_check[f"{TARGET_COLUMN}_dataset"]
    == integrity_check[f"{TARGET_COLUMN}_assignment"]
)

if not target_match.all():
    raise ValueError(
        "Target values do not match between the dataset "
        "and split assignments."
    )

# 5. Attach the previously saved split labels

modeling_df = df.merge(
    split_assignments[["encounter_id", "split"]],
    on="encounter_id",
    how="left",
    validate="one_to_one",
)

if modeling_df["split"].isna().any():
    raise ValueError(
        "Some encounters did not receive a split assignment."
    )

# 6. Create the fixed train, validation, and reserved test sets

train_df = modeling_df[
    modeling_df["split"] == "train"
].copy()

validation_df = modeling_df[
    modeling_df["split"] == "validation"
].copy()

test_df = modeling_df[
    modeling_df["split"] == "test"
].copy()

# Only create train and validation modeling matrices now.
# Test predictors and labels will not be used during development.

X_train = train_df[MODEL_FEATURES].copy()
y_train = train_df[TARGET_COLUMN].copy()

X_validation = validation_df[MODEL_FEATURES].copy()
y_validation = validation_df[TARGET_COLUMN].copy()

# 7. Confirm that patients do not overlap between partitions

train_patients = set(train_df[GROUP_COLUMN])
validation_patients = set(validation_df[GROUP_COLUMN])
test_patients = set(test_df[GROUP_COLUMN])

train_validation_overlap = len(
    train_patients.intersection(validation_patients)
)

train_test_overlap = len(
    train_patients.intersection(test_patients)
)

validation_test_overlap = len(
    validation_patients.intersection(test_patients)
)

if any(
    [
        train_validation_overlap,
        train_test_overlap,
        validation_test_overlap,
    ]
):
    raise ValueError(
        "Patient overlap was detected between saved partitions."
    )

# 8. Display validation results

print("Saved feature and split validation completed successfully.")
print("-" * 65)

print("\nFeature contract:")
print(f"Target column: {TARGET_COLUMN}")
print(f"Patient grouping column: {GROUP_COLUMN}")
print(f"Identifier columns: {IDENTIFIER_COLUMNS}")
print(f"Numeric predictors: {len(NUMERIC_FEATURES)}")
print(f"Categorical predictors: {len(CATEGORICAL_FEATURES)}")
print(f"Total raw predictors: {len(MODEL_FEATURES)}")

print("\nTraining partition:")
print(f"X_train shape: {X_train.shape}")
print(f"Patients: {train_df[GROUP_COLUMN].nunique():,}")
print(f"Positive rate: {y_train.mean():.3%}")

print("\nValidation partition:")
print(f"X_validation shape: {X_validation.shape}")
print(f"Patients: {validation_df[GROUP_COLUMN].nunique():,}")
print(f"Positive rate: {y_validation.mean():.3%}")

print("\nReserved test partition:")
print(f"Rows reserved: {len(test_df):,}")
print(f"Patients reserved: {test_df[GROUP_COLUMN].nunique():,}")
print("Test predictors and labels have not been used for modeling.")

print("\nPatient overlap checks:")
print(
    "Train vs validation:",
    train_validation_overlap,
)
print(
    "Train vs test:",
    train_test_overlap,
)
print(
    "Validation vs test:",
    validation_test_overlap,
)

print("\nSplit assignments were reused without regeneration.")

# Step 3: Rebuild and fit a type-safe preprocessing pipeline

import numpy as np
from scipy import sparse

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_is_fitted
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# 1. Read saved preprocessing settings

RARE_MIN_FREQUENCY = int(
    preprocessing_metadata.get(
        "rare_category_min_frequency",
        10
    )
)

EXPECTED_TRANSFORMED_FEATURES = int(
    preprocessing_metadata.get(
        "transformed_feature_count",
        179
    )
)

metadata_numeric_features = preprocessing_metadata.get(
    "numeric_features"
)

metadata_categorical_features = preprocessing_metadata.get(
    "categorical_features"
)

if (
    metadata_numeric_features is not None
    and list(metadata_numeric_features) != list(NUMERIC_FEATURES)
):
    raise ValueError(
        "Numeric feature order differs between the schema "
        "and preprocessing metadata."
    )

if (
    metadata_categorical_features is not None
    and list(metadata_categorical_features)
    != list(CATEGORICAL_FEATURES)
):
    raise ValueError(
        "Categorical feature order differs between the schema "
        "and preprocessing metadata."
    )

# 2. Define a type-safe rare-category transformer

class RareCategoryGrouper(BaseEstimator, TransformerMixin):
    """
    Converts categorical values to strings and groups categories
    occurring fewer than min_frequency times into __RARE__.

    Categories not observed during fitting are also mapped to
    the rare-category label during transformation.
    """

    def __init__(
        self,
        min_frequency=10,
        rare_label="__RARE__"
    ):
        self.min_frequency = min_frequency
        self.rare_label = rare_label

    def fit(self, X, y=None):
        # Converting all values to strings prevents mixed
        # integer/string categories after inserting __RARE__.
        X_array = np.asarray(X, dtype=object).astype(str)

        if X_array.ndim != 2:
            raise ValueError(
                "RareCategoryGrouper expects a two-dimensional input."
            )

        self.n_features_in_ = X_array.shape[1]
        self.frequent_categories_ = []

        for column_index in range(self.n_features_in_):
            category_counts = pd.Series(
                X_array[:, column_index]
            ).value_counts(dropna=False)

            frequent_categories = set(
                category_counts[
                    category_counts >= self.min_frequency
                ].index.tolist()
            )

            self.frequent_categories_.append(
                frequent_categories
            )

        return self

    def transform(self, X):
        check_is_fitted(
            self,
            attributes=["frequent_categories_"]
        )

        X_array = np.asarray(
            X,
            dtype=object
        ).astype(str)

        if X_array.ndim != 2:
            raise ValueError(
                "RareCategoryGrouper expects a two-dimensional input."
            )

        if X_array.shape[1] != self.n_features_in_:
            raise ValueError(
                "The number of categorical columns differs "
                "from the fitted data."
            )

        for column_index, frequent_categories in enumerate(
            self.frequent_categories_
        ):
            column_values = X_array[:, column_index]

            frequent_mask = np.fromiter(
                (
                    value in frequent_categories
                    for value in column_values
                ),
                dtype=bool,
                count=len(column_values)
            )

            column_values[~frequent_mask] = self.rare_label
            X_array[:, column_index] = column_values

        return X_array

    def get_feature_names_out(self, input_features=None):
        if input_features is None:
            return np.asarray(
                [
                    f"x{index}"
                    for index in range(self.n_features_in_)
                ],
                dtype=object
            )

        return np.asarray(
            input_features,
            dtype=object
        )

# 3. Define numeric preprocessing

numeric_pipeline = Pipeline(
    steps=[
        (
            "median_imputation",
            SimpleImputer(strategy="median")
        ),
        (
            "standard_scaling",
            StandardScaler()
        ),
    ]
)

# 4. Define categorical preprocessing

categorical_pipeline = Pipeline(
    steps=[
        (
            "most_frequent_imputation",
            SimpleImputer(strategy="most_frequent")
        ),
        (
            "rare_category_grouping",
            RareCategoryGrouper(
                min_frequency=RARE_MIN_FREQUENCY,
                rare_label="__RARE__"
            )
        ),
        (
            "one_hot_encoding",
            OneHotEncoder(
                handle_unknown="ignore",
                sparse_output=True
            )
        ),
    ]
)

# 5. Combine numeric and categorical preprocessing

preprocessor = ColumnTransformer(
    transformers=[
        (
            "numeric",
            numeric_pipeline,
            NUMERIC_FEATURES
        ),
        (
            "categorical",
            categorical_pipeline,
            CATEGORICAL_FEATURES
        ),
    ],
    remainder="drop",
    verbose_feature_names_out=True
)

# 6. Fit only on the training partition

X_train_preprocessed = preprocessor.fit_transform(
    X_train,
    y_train
)

# Apply training-fitted preprocessing to validation data
X_validation_preprocessed = preprocessor.transform(
    X_validation
)

# Test data is deliberately not transformed here.

# 7. Retrieve transformed feature names

transformed_feature_names = (
    preprocessor.get_feature_names_out()
)

if (
    len(transformed_feature_names)
    != EXPECTED_TRANSFORMED_FEATURES
):
    raise ValueError(
        "Unexpected transformed feature count. "
        f"Expected {EXPECTED_TRANSFORMED_FEATURES}, "
        f"but found {len(transformed_feature_names)}."
    )

if (
    X_train_preprocessed.shape[1]
    != EXPECTED_TRANSFORMED_FEATURES
):
    raise ValueError(
        "The transformed training matrix has an "
        "unexpected number of columns."
    )

if (
    X_validation_preprocessed.shape[1]
    != EXPECTED_TRANSFORMED_FEATURES
):
    raise ValueError(
        "The transformed validation matrix has an "
        "unexpected number of columns."
    )

# 8. Check for NaN and infinite values

def matrix_contains_only_finite_values(matrix):
    if sparse.issparse(matrix):
        values = matrix.data
    else:
        values = np.asarray(matrix)

    return np.isfinite(values).all()


train_values_valid = matrix_contains_only_finite_values(
    X_train_preprocessed
)

validation_values_valid = matrix_contains_only_finite_values(
    X_validation_preprocessed
)

if not train_values_valid:
    raise ValueError(
        "The transformed training matrix contains "
        "NaN or infinite values."
    )

if not validation_values_valid:
    raise ValueError(
        "The transformed validation matrix contains "
        "NaN or infinite values."
    )

# 9. Display preprocessing results

print("Training-only preprocessing completed successfully.")
print("-" * 65)

print(
    f"\nRare-category minimum frequency: "
    f"{RARE_MIN_FREQUENCY}"
)

print("\nRaw feature counts:")
print(f"Numeric predictors: {len(NUMERIC_FEATURES)}")
print(f"Categorical predictors: {len(CATEGORICAL_FEATURES)}")
print(f"Total raw predictors: {len(MODEL_FEATURES)}")

print("\nTransformed matrices:")
print(
    "Training matrix shape:",
    X_train_preprocessed.shape
)
print(
    "Validation matrix shape:",
    X_validation_preprocessed.shape
)
print(
    "Transformed feature count:",
    len(transformed_feature_names)
)

print("\nMatrix storage:")
print(
    "Training matrix is sparse:",
    sparse.issparse(X_train_preprocessed)
)
print(
    "Validation matrix is sparse:",
    sparse.issparse(X_validation_preprocessed)
)

print("\nValidation checks:")
print(
    "Training values are finite:",
    train_values_valid
)
print(
    "Validation values are finite:",
    validation_values_valid
)

print("\nCategorical type handling:")
print("- All categorical values were converted to strings.")
print("- Numerically coded categorical features remain categorical.")
print("- Rare categories use the string label __RARE__.")

print("\nData-leakage protection:")
print("- Preprocessor was fitted only on X_train.")
print("- Validation data used transform only.")
print("- Test data remains untouched.")

# Step 4: Train and evaluate a majority-class dummy baseline

from sklearn.dummy import DummyClassifier
from sklearn.metrics import (
    accuracy_score,
    balanced_accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score,
    confusion_matrix,
)

# 1. Define a reusable binary-classification evaluation function

def evaluate_binary_classifier(
    model_name,
    y_true,
    y_pred,
    y_probability
):
    """
    Calculate classification metrics for the positive class.
    """

    tn, fp, fn, tp = confusion_matrix(
        y_true,
        y_pred,
        labels=[0, 1]
    ).ravel()

    specificity = (
        tn / (tn + fp)
        if (tn + fp) > 0
        else 0.0
    )

    negative_predictive_value = (
        tn / (tn + fn)
        if (tn + fn) > 0
        else 0.0
    )

    predicted_positive_rate = np.mean(y_pred == 1)

    results = {
        "model": model_name,
        "accuracy": accuracy_score(y_true, y_pred),
        "balanced_accuracy": balanced_accuracy_score(
            y_true,
            y_pred
        ),
        "precision": precision_score(
            y_true,
            y_pred,
            zero_division=0
        ),
        "recall_sensitivity": recall_score(
            y_true,
            y_pred,
            zero_division=0
        ),
        "specificity": specificity,
        "f1_score": f1_score(
            y_true,
            y_pred,
            zero_division=0
        ),
        "roc_auc": roc_auc_score(
            y_true,
            y_probability
        ),
        "pr_auc": average_precision_score(
            y_true,
            y_probability
        ),
        "negative_predictive_value": negative_predictive_value,
        "predicted_positive_rate": predicted_positive_rate,
        "true_negatives": tn,
        "false_positives": fp,
        "false_negatives": fn,
        "true_positives": tp,
    }

    return pd.DataFrame([results])


# 2. Create the majority-class baseline

dummy_majority_model = DummyClassifier(
    strategy="most_frequent",
    random_state=42
)

# 3. Fit using only the training partition

dummy_majority_model.fit(
    X_train_preprocessed,
    y_train
)

# 4. Generate validation predictions

dummy_validation_predictions = (
    dummy_majority_model.predict(
        X_validation_preprocessed
    )
)

positive_class_index = list(
    dummy_majority_model.classes_
).index(1)

dummy_validation_probabilities = (
    dummy_majority_model.predict_proba(
        X_validation_preprocessed
    )[:, positive_class_index]
)

# 5. Evaluate the validation performance

dummy_majority_results = evaluate_binary_classifier(
    model_name="Dummy - Most Frequent Class",
    y_true=y_validation,
    y_pred=dummy_validation_predictions,
    y_probability=dummy_validation_probabilities
)

dummy_confusion_matrix = confusion_matrix(
    y_validation,
    dummy_validation_predictions,
    labels=[0, 1]
)

# 6. Display the results

print("Majority-class dummy baseline completed successfully.")
print("-" * 70)

print("\nModel configuration:")
print("Strategy: Predict the most frequent training class")
print(f"Learned classes: {dummy_majority_model.classes_.tolist()}")
print(
    "Most frequent predicted class:",
    int(dummy_validation_predictions[0])
)

print("\nValidation distribution:")
print(f"Validation encounters: {len(y_validation):,}")
print(f"Actual positive cases: {(y_validation == 1).sum():,}")
print(f"Actual negative cases: {(y_validation == 0).sum():,}")
print(f"Actual positive rate: {y_validation.mean():.3%}")

print("\nPrediction distribution:")
print(
    "Predicted positive cases:",
    int((dummy_validation_predictions == 1).sum())
)
print(
    "Predicted negative cases:",
    int((dummy_validation_predictions == 0).sum())
)
print(
    "Predicted positive rate:",
    f"{np.mean(dummy_validation_predictions == 1):.3%}"
)

print("\nConfusion matrix:")
print("Rows = actual class; columns = predicted class")
print(dummy_confusion_matrix)

print("\nValidation metrics:")
print(
    dummy_majority_results[
        [
            "model",
            "accuracy",
            "balanced_accuracy",
            "precision",
            "recall_sensitivity",
            "specificity",
            "f1_score",
            "roc_auc",
            "pr_auc",
        ]
    ].round(4).to_string(index=False)
)

print("\nTest set status:")
print("- Test data was not transformed.")
print("- Test data was not used for fitting.")
print("- Test data was not used for evaluation.")

#plot and save dummy baseline confusion matrix

import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import ConfusionMatrixDisplay

# Create output folder 
FIGURES_PATH = PROJECT_ROOT / "outputs" / "figures"
FIGURES_PATH.mkdir(parents=True, exist_ok=True)

# Plot confusion matrix
fig, ax = plt.subplots(figsize=(6, 5))

disp = ConfusionMatrixDisplay(
    confusion_matrix=dummy_confusion_matrix,
    display_labels=["Not Readmitted (0)", "Readmitted (1)"]
)

disp.plot(
    cmap="Blues",
    values_format="d",
    ax=ax,
    colorbar=False
)

ax.set_title("Dummy Baseline Confusion Matrix (Validation Set)")
plt.tight_layout()

# Save figure
save_path = FIGURES_PATH / "dummy_baseline_confusion_matrix.png"
plt.savefig(save_path, dpi=300, bbox_inches="tight")
plt.show()

print(f"Figure saved to: {save_path}")

# Step 5: Train and evaluate unweighted logistic regression

import time
import warnings
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.exceptions import ConvergenceWarning
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 1. Define the baseline logistic regression model

logistic_baseline_model = LogisticRegression(
    penalty="l2",
    C=1.0,
    class_weight=None,
    solver="liblinear",
    max_iter=1000,
    random_state=42
)

# 2. Fit using training data only

training_start_time = time.perf_counter()

with warnings.catch_warnings():
    warnings.simplefilter("error", ConvergenceWarning)

    logistic_baseline_model.fit(
        X_train_preprocessed,
        y_train
    )

training_time_seconds = (
    time.perf_counter() - training_start_time
)

# 3. Generate validation predictions

logistic_validation_predictions = (
    logistic_baseline_model.predict(
        X_validation_preprocessed
    )
)

positive_class_index = list(
    logistic_baseline_model.classes_
).index(1)

logistic_validation_probabilities = (
    logistic_baseline_model.predict_proba(
        X_validation_preprocessed
    )[:, positive_class_index]
)

# 4. Evaluate validation performance

logistic_baseline_results = evaluate_binary_classifier(
    model_name="Logistic Regression - Unweighted",
    y_true=y_validation,
    y_pred=logistic_validation_predictions,
    y_probability=logistic_validation_probabilities
)

logistic_baseline_results[
    "training_time_seconds"
] = training_time_seconds

logistic_baseline_results[
    "iterations"
] = int(logistic_baseline_model.n_iter_[0])

logistic_confusion_matrix = confusion_matrix(
    y_validation,
    logistic_validation_predictions,
    labels=[0, 1]
)

# 5. Display confusion matrix as an image

fig, ax = plt.subplots(figsize=(6, 5))

display = ConfusionMatrixDisplay(
    confusion_matrix=logistic_confusion_matrix,
    display_labels=[
        "Not Readmitted (0)",
        "Readmitted (1)"
    ]
)

display.plot(
    cmap="Blues",
    values_format="d",
    colorbar=False,
    ax=ax
)

ax.set_title(
    "Unweighted Logistic Regression\n"
    "Validation Confusion Matrix"
)

plt.tight_layout()
plt.show()

# 6. Display compact validation results

print("Unweighted Logistic Regression — Validation Results")
print("-" * 65)

print("\nModel configuration:")
print("Penalty: L2")
print("C: 1.0")
print("Class weight: None")
print("Solver: liblinear")
print("Decision threshold: 0.50")
print(
    f"Iterations used: "
    f"{logistic_baseline_model.n_iter_[0]}"
)
print(
    f"Training time: "
    f"{training_time_seconds:.2f} seconds"
)

print("\nConfusion matrix:")
print("Rows = actual [0, 1]; columns = predicted [0, 1]")
print(logistic_confusion_matrix)

print("\nValidation metrics:")

display_columns = [
    "accuracy",
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "specificity",
    "f1_score",
    "roc_auc",
    "pr_auc",
    "predicted_positive_rate",
    "true_negatives",
    "false_positives",
    "false_negatives",
    "true_positives",
]

print(
    logistic_baseline_results[
        display_columns
    ]
    .round(4)
    .T
    .to_string(header=False)
)

# 7. Compare with the dummy baseline

baseline_model_comparison = pd.concat(
    [
        dummy_majority_results,
        logistic_baseline_results
    ],
    ignore_index=True
)

comparison_columns = [
    "model",
    "accuracy",
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "specificity",
    "f1_score",
    "roc_auc",
    "pr_auc",
]

print("\nComparison with dummy baseline:")

print(
    baseline_model_comparison[
        comparison_columns
    ]
    .round(4)
    .to_string(index=False)
)

print("\nTest set status:")
print("- Test data remains untouched.")
print("- Evaluation used validation data only.")

# Step 6: Train class-weighted logistic regression

import time
import warnings
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.exceptions import ConvergenceWarning
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 1. Define the class-weighted logistic regression model

weighted_logistic_model = LogisticRegression(
    penalty="l2",
    C=1.0,
    class_weight="balanced",
    solver="liblinear",
    max_iter=1000,
    random_state=42
)

# 2. Fit using training data only

training_start_time = time.perf_counter()

with warnings.catch_warnings():
    warnings.simplefilter("error", ConvergenceWarning)

    weighted_logistic_model.fit(
        X_train_preprocessed,
        y_train
    )

weighted_training_time_seconds = (
    time.perf_counter() - training_start_time
)

# 3. Generate validation predictions and probabilities

weighted_logistic_predictions = (
    weighted_logistic_model.predict(
        X_validation_preprocessed
    )
)

weighted_positive_class_index = list(
    weighted_logistic_model.classes_
).index(1)

weighted_logistic_probabilities = (
    weighted_logistic_model.predict_proba(
        X_validation_preprocessed
    )[:, weighted_positive_class_index]
)

# 4. Evaluate validation performance

weighted_logistic_results = evaluate_binary_classifier(
    model_name="Logistic Regression - Class Weighted",
    y_true=y_validation,
    y_pred=weighted_logistic_predictions,
    y_probability=weighted_logistic_probabilities
)

weighted_logistic_results[
    "training_time_seconds"
] = weighted_training_time_seconds

weighted_logistic_results[
    "iterations"
] = int(weighted_logistic_model.n_iter_[0])

weighted_logistic_confusion_matrix = confusion_matrix(
    y_validation,
    weighted_logistic_predictions,
    labels=[0, 1]
)

# 5. Display confusion matrix image

fig, ax = plt.subplots(figsize=(6, 5))

weighted_display = ConfusionMatrixDisplay(
    confusion_matrix=weighted_logistic_confusion_matrix,
    display_labels=[
        "Not Readmitted (0)",
        "Readmitted (1)"
    ]
)

weighted_display.plot(
    cmap="Blues",
    values_format="d",
    colorbar=False,
    ax=ax
)

ax.set_title(
    "Class-Weighted Logistic Regression\n"
    "Validation Confusion Matrix"
)

plt.tight_layout()
plt.show()

# 6. Display compact validation results

print("Class-Weighted Logistic Regression — Validation Results")
print("-" * 70)

print("\nModel configuration:")
print("Penalty: L2")
print("C: 1.0")
print("Class weight: balanced")
print("Solver: liblinear")
print("Decision threshold: 0.50")
print(
    f"Iterations used: "
    f"{weighted_logistic_model.n_iter_[0]}"
)
print(
    f"Training time: "
    f"{weighted_training_time_seconds:.2f} seconds"
)

print("\nConfusion matrix:")
print("Rows = actual [0, 1]; columns = predicted [0, 1]")
print(weighted_logistic_confusion_matrix)

print("\nValidation metrics:")

display_columns = [
    "accuracy",
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "specificity",
    "f1_score",
    "roc_auc",
    "pr_auc",
    "predicted_positive_rate",
    "true_negatives",
    "false_positives",
    "false_negatives",
    "true_positives",
]

print(
    weighted_logistic_results[
        display_columns
    ]
    .round(4)
    .T
    .to_string(header=False)
)

# 7. Compare all models completed so far

baseline_model_comparison = pd.concat(
    [
        dummy_majority_results,
        logistic_baseline_results,
        weighted_logistic_results
    ],
    ignore_index=True
)

comparison_columns = [
    "model",
    "accuracy",
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "specificity",
    "f1_score",
    "roc_auc",
    "pr_auc",
]

print("\nModel comparison:")

print(
    baseline_model_comparison[
        comparison_columns
    ]
    .round(4)
    .to_string(index=False)
)

print("\nTest set status:")
print("- Test data remains untouched.")
print("- Evaluation used validation data only.")

# Step 7: Train a controlled class-weighted decision tree

import time
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# 1. Define a controlled decision tree

decision_tree_model = DecisionTreeClassifier(
    criterion="gini",
    max_depth=6,
    min_samples_split=100,
    min_samples_leaf=50,
    class_weight="balanced",
    random_state=42
)

# 2. Fit using training data only

training_start_time = time.perf_counter()

decision_tree_model.fit(
    X_train_preprocessed,
    y_train
)

decision_tree_training_time = (
    time.perf_counter() - training_start_time
)

# 3. Generate validation predictions and probabilities

decision_tree_predictions = decision_tree_model.predict(
    X_validation_preprocessed
)

decision_tree_positive_class_index = list(
    decision_tree_model.classes_
).index(1)

decision_tree_probabilities = (
    decision_tree_model.predict_proba(
        X_validation_preprocessed
    )[:, decision_tree_positive_class_index]
)

# 4. Evaluate validation performance

decision_tree_results = evaluate_binary_classifier(
    model_name="Decision Tree - Class Weighted",
    y_true=y_validation,
    y_pred=decision_tree_predictions,
    y_probability=decision_tree_probabilities
)

decision_tree_results[
    "training_time_seconds"
] = decision_tree_training_time

decision_tree_results[
    "tree_depth"
] = decision_tree_model.get_depth()

decision_tree_results[
    "number_of_leaves"
] = decision_tree_model.get_n_leaves()

decision_tree_confusion_matrix = confusion_matrix(
    y_validation,
    decision_tree_predictions,
    labels=[0, 1]
)

# 5. Display confusion matrix image

fig, ax = plt.subplots(figsize=(6, 5))

tree_display = ConfusionMatrixDisplay(
    confusion_matrix=decision_tree_confusion_matrix,
    display_labels=[
        "Not Readmitted (0)",
        "Readmitted (1)"
    ]
)

tree_display.plot(
    cmap="Blues",
    values_format="d",
    colorbar=False,
    ax=ax
)

ax.set_title(
    "Class-Weighted Decision Tree\n"
    "Validation Confusion Matrix"
)

plt.tight_layout()
plt.show()

# 6. Display compact validation results

print("Class-Weighted Decision Tree — Validation Results")
print("-" * 65)

print("\nModel configuration:")
print("Criterion: gini")
print("Maximum depth: 6")
print("Minimum samples per split: 100")
print("Minimum samples per leaf: 50")
print("Class weight: balanced")
print("Decision threshold: 0.50")

print("\nFitted tree structure:")
print(f"Actual tree depth: {decision_tree_model.get_depth()}")
print(f"Number of leaves: {decision_tree_model.get_n_leaves()}")
print(f"Training time: {decision_tree_training_time:.2f} seconds")

print("\nConfusion matrix:")
print("Rows = actual [0, 1]; columns = predicted [0, 1]")
print(decision_tree_confusion_matrix)

print("\nValidation metrics:")

display_columns = [
    "accuracy",
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "specificity",
    "f1_score",
    "roc_auc",
    "pr_auc",
    "predicted_positive_rate",
    "true_negatives",
    "false_positives",
    "false_negatives",
    "true_positives",
]

print(
    decision_tree_results[
        display_columns
    ]
    .round(4)
    .T
    .to_string(header=False)
)

# 7. Compare all completed baseline models

baseline_model_comparison = pd.concat(
    [
        dummy_majority_results,
        logistic_baseline_results,
        weighted_logistic_results,
        decision_tree_results
    ],
    ignore_index=True
)

comparison_columns = [
    "model",
    "accuracy",
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "specificity",
    "f1_score",
    "roc_auc",
    "pr_auc",
]

print("\nBaseline model comparison:")

print(
    baseline_model_comparison[
        comparison_columns
    ]
    .round(4)
    .to_string(index=False)
)

print("\nTest set status:")
print("- Test data remains untouched.")
print("- Evaluation used validation data only.")

# Step 8: Compare and save all baseline-model results

from pathlib import Path

# 1. Combine the completed baseline results

baseline_model_comparison = pd.concat(
    [
        dummy_majority_results,
        logistic_baseline_results,
        weighted_logistic_results,
        decision_tree_results,
    ],
    ignore_index=True
)

# 2. Select and organize comparison columns

comparison_columns = [
    "model",
    "accuracy",
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "specificity",
    "f1_score",
    "roc_auc",
    "pr_auc",
    "predicted_positive_rate",
    "true_negatives",
    "false_positives",
    "false_negatives",
    "true_positives",
]

baseline_model_comparison = (
    baseline_model_comparison[comparison_columns]
    .sort_values(
        by="balanced_accuracy",
        ascending=False
    )
    .reset_index(drop=True)
)

# 3. Display a compact percentage comparison

percentage_columns = [
    "accuracy",
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "specificity",
    "f1_score",
    "roc_auc",
    "pr_auc",
    "predicted_positive_rate",
]

display_comparison = baseline_model_comparison.copy()

display_comparison[percentage_columns] = (
    display_comparison[percentage_columns] * 100
)

print("Baseline Model Comparison — Validation Set")
print("-" * 100)

print(
    display_comparison[
        [
            "model",
            "accuracy",
            "balanced_accuracy",
            "precision",
            "recall_sensitivity",
            "specificity",
            "f1_score",
            "roc_auc",
            "pr_auc",
        ]
    ]
    .round(2)
    .to_string(index=False)
)

print("\nConfusion-matrix counts:")
print(
    baseline_model_comparison[
        [
            "model",
            "true_negatives",
            "false_positives",
            "false_negatives",
            "true_positives",
        ]
    ]
    .astype(
        {
            "true_negatives": int,
            "false_positives": int,
            "false_negatives": int,
            "true_positives": int,
        }
    )
    .to_string(index=False)
)

# 4. Identify metric leaders

metrics_to_rank = [
    "balanced_accuracy",
    "recall_sensitivity",
    "f1_score",
    "roc_auc",
    "pr_auc",
]

print("\nValidation metric leaders:")

for metric in metrics_to_rank:
    best_row = baseline_model_comparison.loc[
        baseline_model_comparison[metric].idxmax()
    ]

    print(
        f"- {metric}: {best_row['model']} "
        f"({best_row[metric]:.4f})"
    )

# 5. Save the comparison table

METRICS_PATH = PROJECT_ROOT / "outputs" / "metrics"
METRICS_PATH.mkdir(parents=True, exist_ok=True)

comparison_save_path = (
    METRICS_PATH
    / "notebook_4_baseline_model_comparison.csv"
)

baseline_model_comparison.to_csv(
    comparison_save_path,
    index=False
)

print("\nSaved output:")
print(comparison_save_path)

print("\nTest set status:")
print("- Test data remains untouched.")
print("- All comparisons use validation results only.")

# Step 9: Create validation ROC and precision-recall curves

import matplotlib.pyplot as plt

from sklearn.metrics import (
    roc_curve,
    roc_auc_score,
    precision_recall_curve,
    average_precision_score,
)

# 1. Collect validation probabilities from completed models

baseline_probability_results = {
    "Dummy Classifier": dummy_validation_probabilities,
    "Unweighted Logistic Regression":
        logistic_validation_probabilities,
    "Class-Weighted Logistic Regression":
        weighted_logistic_probabilities,
    "Class-Weighted Decision Tree":
        decision_tree_probabilities,
}

FIGURES_PATH = PROJECT_ROOT / "outputs" / "figures"
FIGURES_PATH.mkdir(parents=True, exist_ok=True)

# 2. Plot ROC curves

fig, ax = plt.subplots(figsize=(8, 6))

for model_name, probabilities in baseline_probability_results.items():

    false_positive_rate, true_positive_rate, _ = roc_curve(
        y_validation,
        probabilities
    )

    model_roc_auc = roc_auc_score(
        y_validation,
        probabilities
    )

    ax.plot(
        false_positive_rate,
        true_positive_rate,
        linewidth=2,
        label=f"{model_name} (AUC = {model_roc_auc:.3f})"
    )

# Random-classifier reference line
ax.plot(
    [0, 1],
    [0, 1],
    linestyle="--",
    linewidth=1.5,
    label="Random Classifier (AUC = 0.500)"
)

ax.set_title("Baseline Models — Validation ROC Curves")
ax.set_xlabel("False Positive Rate")
ax.set_ylabel("True Positive Rate")
ax.legend(loc="lower right")
ax.grid(alpha=0.25)

plt.tight_layout()

roc_save_path = (
    FIGURES_PATH
    / "notebook_4_baseline_roc_curves.png"
)

plt.savefig(
    roc_save_path,
    dpi=300,
    bbox_inches="tight"
)

plt.show()

# 3. Plot precision-recall curves

fig, ax = plt.subplots(figsize=(8, 6))

for model_name, probabilities in baseline_probability_results.items():

    precision_values, recall_values, _ = precision_recall_curve(
        y_validation,
        probabilities
    )

    model_pr_auc = average_precision_score(
        y_validation,
        probabilities
    )

    ax.plot(
        recall_values,
        precision_values,
        linewidth=2,
        label=f"{model_name} (AP = {model_pr_auc:.3f})"
    )

# Positive-class prevalence reference
validation_prevalence = y_validation.mean()

ax.axhline(
    y=validation_prevalence,
    linestyle="--",
    linewidth=1.5,
    label=(
        "Positive-Class Prevalence "
        f"({validation_prevalence:.3f})"
    )
)

ax.set_title(
    "Baseline Models — Validation Precision-Recall Curves"
)
ax.set_xlabel("Recall")
ax.set_ylabel("Precision")
ax.legend(loc="upper right")
ax.grid(alpha=0.25)

plt.tight_layout()

pr_save_path = (
    FIGURES_PATH
    / "notebook_4_baseline_precision_recall_curves.png"
)

plt.savefig(
    pr_save_path,
    dpi=300,
    bbox_inches="tight"
)

plt.show()

# 4. Confirm saved figures

print("Baseline validation curves created successfully.")
print("-" * 65)

print("\nSaved ROC curve:")
print(roc_save_path)

print("\nSaved precision-recall curve:")
print(pr_save_path)

print("\nValidation prevalence:")
print(f"{validation_prevalence:.4f} ({validation_prevalence:.2%})")

print("\nTest set status:")
print("- Test data remains untouched.")
print("- Curves use validation predictions only.")

# Step 10: Create baseline validation metric comparison chart

import matplotlib.pyplot as plt
import pandas as pd

# 1. Select the most important comparison metrics

chart_metrics = [
    "balanced_accuracy",
    "precision",
    "recall_sensitivity",
    "f1_score",
    "roc_auc",
    "pr_auc",
]

model_label_map = {
    "Dummy - Most Frequent Class": "Dummy",
    "Logistic Regression - Unweighted": "Unweighted Logistic",
    "Logistic Regression - Class Weighted": "Weighted Logistic",
    "Decision Tree - Class Weighted": "Weighted Decision Tree",
}

chart_data = baseline_model_comparison[
    ["model"] + chart_metrics
].copy()

chart_data["model"] = (
    chart_data["model"]
    .map(model_label_map)
    .fillna(chart_data["model"])
)

# Convert metrics to percentages for easier interpretation
chart_data[chart_metrics] = (
    chart_data[chart_metrics] * 100
)

chart_data = chart_data.set_index("model")

# 2. Rename metrics for display

chart_data = chart_data.rename(
    columns={
        "balanced_accuracy": "Balanced Accuracy",
        "precision": "Precision",
        "recall_sensitivity": "Recall",
        "f1_score": "F1 Score",
        "roc_auc": "ROC-AUC",
        "pr_auc": "PR-AUC",
    }
)

# 3. Create the grouped bar chart

ax = chart_data.plot(
    kind="bar",
    figsize=(12, 7),
    width=0.82
)

ax.set_title(
    "Baseline Models — Validation Metric Comparison",
    fontsize=15
)

ax.set_xlabel("Model")
ax.set_ylabel("Metric Value (%)")
ax.set_ylim(0, 100)

ax.tick_params(
    axis="x",
    rotation=15
)

for label in ax.get_xticklabels():
    label.set_horizontalalignment("right")

ax.legend(
    title="Validation Metric",
    bbox_to_anchor=(1.02, 1),
    loc="upper left"
)

ax.grid(
    axis="y",
    alpha=0.25
)

plt.tight_layout()

# 4. Save and display the chart

comparison_chart_path = (
    FIGURES_PATH
    / "notebook_4_baseline_metric_comparison.png"
)

plt.savefig(
    comparison_chart_path,
    dpi=300,
    bbox_inches="tight"
)

plt.show()

# 5. Confirm the output

print("Baseline metric comparison chart created successfully.")
print("-" * 65)

print("\nSaved figure:")
print(comparison_chart_path)

print("\nModels included:")
for model_name in chart_data.index:
    print(f"- {model_name}")

print("\nMetrics included:")
for metric_name in chart_data.columns:
    print(f"- {metric_name}")

print("\nTest set status:")
print("- Test data remains untouched.")
print("- Chart uses validation results only.")

# Step 11: Save Notebook 4 models and completion metadata

import json
import joblib
from datetime import datetime

# 1. Create output directories

MODELS_PATH = PROJECT_ROOT / "models"
ARTIFACTS_PATH = PROJECT_ROOT / "artifacts"
METRICS_PATH = PROJECT_ROOT / "outputs" / "metrics"

MODELS_PATH.mkdir(parents=True, exist_ok=True)
ARTIFACTS_PATH.mkdir(parents=True, exist_ok=True)
METRICS_PATH.mkdir(parents=True, exist_ok=True)

# 2. Define saved artifact paths

preprocessor_save_path = (
    MODELS_PATH
    / "notebook_4_fitted_preprocessor.joblib"
)

dummy_model_save_path = (
    MODELS_PATH
    / "notebook_4_dummy_classifier.joblib"
)

unweighted_logistic_save_path = (
    MODELS_PATH
    / "notebook_4_logistic_unweighted.joblib"
)

weighted_logistic_save_path = (
    MODELS_PATH
    / "notebook_4_logistic_class_weighted.joblib"
)

decision_tree_save_path = (
    MODELS_PATH
    / "notebook_4_decision_tree_class_weighted.joblib"
)

metadata_save_path = (
    ARTIFACTS_PATH
    / "notebook_4_baseline_metadata.json"
)

completion_summary_path = (
    METRICS_PATH
    / "notebook_4_completion_summary.csv"
)

# 3. Save the fitted preprocessor and trained models

joblib.dump(
    preprocessor,
    preprocessor_save_path
)

joblib.dump(
    dummy_majority_model,
    dummy_model_save_path
)

joblib.dump(
    logistic_baseline_model,
    unweighted_logistic_save_path
)

joblib.dump(
    weighted_logistic_model,
    weighted_logistic_save_path
)

joblib.dump(
    decision_tree_model,
    decision_tree_save_path
)

# 4. Identify the provisional leading baseline

leading_baseline_row = (
    baseline_model_comparison
    .sort_values(
        by=[
            "balanced_accuracy",
            "f1_score",
            "roc_auc",
            "pr_auc"
        ],
        ascending=False
    )
    .iloc[0]
)

leading_baseline_name = leading_baseline_row["model"]

# 5. Create reproducibility metadata

baseline_metadata = {
    "notebook": "04_baseline_models.ipynb",
    "created_at": datetime.now().isoformat(timespec="seconds"),
    "target_column": TARGET_COLUMN,
    "group_column": GROUP_COLUMN,
    "identifier_columns": IDENTIFIER_COLUMNS,
    "numeric_feature_count": len(NUMERIC_FEATURES),
    "categorical_feature_count": len(CATEGORICAL_FEATURES),
    "raw_feature_count": len(MODEL_FEATURES),
    "transformed_feature_count": len(transformed_feature_names),
    "rare_category_min_frequency": RARE_MIN_FREQUENCY,
    "train_encounters": int(len(train_df)),
    "validation_encounters": int(len(validation_df)),
    "reserved_test_encounters": int(len(test_df)),
    "models_evaluated": baseline_model_comparison[
        "model"
    ].tolist(),
    "provisional_leading_baseline": leading_baseline_name,
    "selection_basis": [
        "balanced_accuracy",
        "f1_score",
        "roc_auc",
        "pr_auc",
        "recall_sensitivity"
    ],
    "test_set_used": False,
    "saved_files": {
        "fitted_preprocessor": str(preprocessor_save_path),
        "dummy_classifier": str(dummy_model_save_path),
        "unweighted_logistic_regression": str(
            unweighted_logistic_save_path
        ),
        "class_weighted_logistic_regression": str(
            weighted_logistic_save_path
        ),
        "class_weighted_decision_tree": str(
            decision_tree_save_path
        ),
        "comparison_table": str(comparison_save_path),
        "roc_curve": str(roc_save_path),
        "precision_recall_curve": str(pr_save_path),
        "metric_comparison_chart": str(
            comparison_chart_path
        )
    }
}

with open(
    metadata_save_path,
    "w",
    encoding="utf-8"
) as file:
    json.dump(
        baseline_metadata,
        file,
        indent=4
    )

# 6. Save a compact notebook completion summary

completion_summary = pd.DataFrame(
    [
        {
            "notebook": "04_baseline_models.ipynb",
            "status": "Complete",
            "models_evaluated": len(
                baseline_model_comparison
            ),
            "leading_baseline": leading_baseline_name,
            "leading_balanced_accuracy": (
                leading_baseline_row[
                    "balanced_accuracy"
                ]
            ),
            "leading_recall": (
                leading_baseline_row[
                    "recall_sensitivity"
                ]
            ),
            "leading_f1_score": (
                leading_baseline_row[
                    "f1_score"
                ]
            ),
            "leading_roc_auc": (
                leading_baseline_row[
                    "roc_auc"
                ]
            ),
            "leading_pr_auc": (
                leading_baseline_row[
                    "pr_auc"
                ]
            ),
            "test_set_used": False
        }
    ]
)

completion_summary.to_csv(
    completion_summary_path,
    index=False
)

# 7. Confirm that all files were saved

saved_paths = [
    preprocessor_save_path,
    dummy_model_save_path,
    unweighted_logistic_save_path,
    weighted_logistic_save_path,
    decision_tree_save_path,
    metadata_save_path,
    comparison_save_path,
    completion_summary_path,
    roc_save_path,
    pr_save_path,
    comparison_chart_path,
]

missing_saved_files = [
    str(path)
    for path in saved_paths
    if not path.exists()
]

if missing_saved_files:
    raise FileNotFoundError(
        "Some Notebook 4 outputs were not saved:\n"
        + "\n".join(missing_saved_files)
    )

# 8. Display completion information

print("Notebook 4 artifacts saved successfully.")
print("-" * 70)

print("\nProvisional leading baseline:")
print(leading_baseline_name)

print("\nLeading baseline validation results:")
print(
    f"Balanced accuracy: "
    f"{leading_baseline_row['balanced_accuracy']:.4f}"
)
print(
    f"Recall: "
    f"{leading_baseline_row['recall_sensitivity']:.4f}"
)
print(
    f"F1 score: "
    f"{leading_baseline_row['f1_score']:.4f}"
)
print(
    f"ROC-AUC: "
    f"{leading_baseline_row['roc_auc']:.4f}"
)
print(
    f"PR-AUC: "
    f"{leading_baseline_row['pr_auc']:.4f}"
)

print("\nSaved model files:")
print(f"- {preprocessor_save_path}")
print(f"- {dummy_model_save_path}")
print(f"- {unweighted_logistic_save_path}")
print(f"- {weighted_logistic_save_path}")
print(f"- {decision_tree_save_path}")

print("\nSaved documentation files:")
print(f"- {metadata_save_path}")
print(f"- {comparison_save_path}")
print(f"- {completion_summary_path}")

print("\nNotebook 4 status:")
print("- Baseline modeling completed.")
print("- Validation comparisons completed.")
print("- Models and preprocessing artifacts saved.")
print("- Test set remains completely untouched.")

```
